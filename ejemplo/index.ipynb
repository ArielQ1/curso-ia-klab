```python
import pandas as pd
import numpy as np

datos = pd.read_csv('./Iris.csv')
for columna in datos.select_dtypes(include="number").columns:
    print(f" ------- Columna: {columna}")
    media = np.mean(datos[columna])
    mediana = np.median(datos[columna])
    desviacion = np.std(datos[columna])
    print(f"Media: {media}")
    print(f"Mediana: {mediana}")
    print(f"Desviación estándar: {desviacion}")

```

```python
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, label='Seno')
plt.plot(x, y_cos, label='Coseno')
plt.title('Gráfica de Seno y Coseno')
plt.xlabel('Ángulo (radianes)')
plt.ylabel('Valor')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

datos_encuesta = pd.read_csv('./datos.csv')
datos_encuesta['edad'] = datos_encuesta['edad'].astype(int)
datos_encuesta['edad'].hist(bins=10)
plt.title('Histograma de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
print(datos_encuesta['cabello'].value_counts())
datos_encuesta['cabello'].hist(bins=10)
plt.title('Histograma de cabello')
plt.xlabel('Cabello')
plt.ylabel('Frecuencia')
plt.show()
```

```python
datos2 = [[1, 5, np.nan, 6, 7], [np.nan, np.nan, 8, 9, 10], [3, 4, 5, 6, np.nan], [np.nan, 2, 1, 0, 3]]
datos2 = np.array(datos2)
print(datos2)
```

```python
from sklearn.impute import SimpleImputer
pre_procesamiento = SimpleImputer(missing_values=np.nan, strategy='mean')
datos_pre_procesados = pre_procesamiento.fit_transform(datos2)

print(datos_pre_procesados)
```

```python
from sklearn.preprocessing import Normalizer
pre_procesamiento = Normalizer(norm='l2')
datos_normalizados = pre_procesamiento.fit_transform(datos_pre_procesados)
print(datos_normalizados)
```
